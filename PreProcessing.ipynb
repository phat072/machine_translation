```python
from bpemb import BPEmb

bpemb_vi = BPEmb(lang="vi", vs=50000, dim=200)
bpemb_zh = BPEmb(lang="zh", vs=50000, dim=200)
```

```python
def tokenizer(text, lang):
  if lang == "vi":
    text_pp = " ".join(bpemb_vi.encode(text))
  else:
    text_pp = " ".join(bpemb_zh.encode(text))

  return text_pp
```

```python
tokenizer(df_train["vi"][0], "vi")
```

```python
df_train["vi_token"] = df_train["vi"].apply(lambda x: tokenizer(x, "vi"))
df_train["zh_token"] = df_train["zh"].apply(lambda x: tokenizer(x, "zh"))
```

# **Pre processing**

## **Punc norm**


*   – to -



```python
!pip install sacremoses
```

```python
from sacremoses import MosesPunctNormalizer

mpn_vi = MosesPunctNormalizer(lang = "vi")
mpn_zh = MosesPunctNormalizer(lang = "zh")

def punc_norm(text, lang):
  if lang == "vi":
    return mpn_vi.normalize(text)

  return mpn_zh.normalize(text)
```

```python
df_train["vi_punc"] = df_train["vi"].apply(lambda x: punc_norm(x, "vi"))
df_train["zh_punc"] = df_train["zh"].apply(lambda x: punc_norm(x, "zh"))

df_train
```

```python
df_test["vi_punc"] = df_test["vi"].apply(lambda x: punc_norm(x, "vi"))
df_test["zh_punc"] = df_test["zh"].apply(lambda x: punc_norm(x, "zh"))

df_test
```

## **Tokenization**

```python
df_train["vi_token"] = df_train["vi_punc"].apply(lambda x: tokenizer(x, "vi"))
df_train["zh_token"] = df_train["zh_punc"].apply(lambda x: tokenizer(x, "zh"))

df_train
```

```python
df_test["vi_token"] = df_test["vi_punc"].apply(lambda x: tokenizer(x, "vi"))
df_test["zh_token"] = df_test["zh_punc"].apply(lambda x: tokenizer(x, "zh"))

df_test
```

```python
df_train.to_csv(path_pre + "train.csv", index = False)
df_test.to_csv(path_pre + "test.csv", index = False)
```

## **Cleaning and Filtering**


*   Deduplication
*   Empty or mostly containing non-printable characters.
*   With invalid UTF8, i.e., not properly encoded.
*   Containing extremely long tokens (or “words”) since they are often        non-translatable, for instance, DNA sequences, digit sequences, nonsense, etc.
*   And, sometimes, duplicates, i.e., if a segment, or a pair of segments, appear more than once in the parallel data, we keep only one instance of it.
