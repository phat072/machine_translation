```python
max_length_i = max(len(i) for i in X_train)
max_length_o = max(len(i) for i in y_train)
print(max_length_i, max_length_o)
```

```python
from torch.utils.data import Dataset, DataLoader

class CustomDataset(Dataset):
    def __init__ (self, tokenizer, X_data, y_data):
        self.tokenizer = tokenizer
        self.X_data = X_data
        self.y_data = y_data
    def __len__ (self):
        return len(self.X_data)
    def __getitem__(self, idx):
        source = self.tokenizer.encode_plus(
            "translate Vietnamese to Chinese: " + self.X_data[idx],
            max_length=max_length_i,
            padding="max_length",
            pad_to_max_length=True,
            truncation=True,
            return_tensors="pt"
        )

        with self.tokenizer.as_target_tokenizer():
          target = self.tokenizer.encode_plus(
              self.y_data[idx],
              max_length=max_length_o,
              padding="max_length",
              pad_to_max_length=True,
              truncation=True,
              return_tensors="pt"
          )

        source_ids = source['input_ids'].squeeze()
        source_mask = source['attention_mask'].squeeze()
        target_ids = target['input_ids'].squeeze()
        target_mask = target['attention_mask'].squeeze()

        return {
            'input_ids': source_ids.to(dtype=torch.long), 
            'attention_mask': source_mask.to(dtype=torch.long), 
            'labels': target_ids.to(dtype=torch.long),
            # 'decoder_attention_mask': target_mask.to(dtype=torch.long)
        }
```

```python
from transformers.trainer_pt_utils import get_parameter_names


def get_optimizer(model, train_args, type = "AdamW"):
  decay_parameters = get_parameter_names(model, [torch.nn.LayerNorm])
  decay_parameters = [name for name in decay_parameters if "bias" not in name]
  optimizer_grouped_parameters = [
      {
          "params": [p for n, p in model.named_parameters() if n in decay_parameters],
          "weight_decay": train_args.weight_decay,
      },
      {
          "params": [p for n, p in model.named_parameters() if n not in decay_parameters],
          "weight_decay": 0.0,
      },
  ]
  optimizer = torch.optim.AdamW(
      optimizer_grouped_parameters,
      lr=train_args.learning_rate,
      betas=(train_args.adam_beta1, train_args.adam_beta2),
      eps=train_args.adam_epsilon
  )

  return optimizer
```

```python
from transformers import AutoModelForSeq2SeqLM, DataCollatorForSeq2Seq, Seq2SeqTrainingArguments, Seq2SeqTrainer, AutoTokenizer


name = "google/mt5-small"
name = "t5-small"
# name = "gsarti/it5-small"


tokenizer = AutoTokenizer.from_pretrained(name)
model = AutoModelForSeq2SeqLM.from_pretrained(name).to(device)
```

```python
train_dataset = CustomDataset(tokenizer, X_train, y_train)
test_dataset = CustomDataset(tokenizer, X_test, y_test)
```

```python
idx = 0
print(X_train[idx])
print(tokenizer.decode(train_dataset[idx]['input_ids'], skip_special_tokens=True))
```

```python
epochs = 1
batch_size = 16
learning_rate = 3e-4
weight_decay = 0.01

free_gpu()
```

```python
training_args = Seq2SeqTrainingArguments(
    output_dir = '/content/drive/MyDrive/C02.EMit-Evalita2023/Model/',
    # evaluation_strategy= 'epoch',
    learning_rate = learning_rate,
    save_strategy = 'no',
    per_device_train_batch_size = batch_size,
    num_train_epochs = epochs,
    weight_decay = weight_decay,
    predict_with_generate=True,
    fp16=True,
    fp16_full_eval=True,
    jit_mode_eval=True,
    logging_strategy="steps",
    logging_steps = 100,
    seed = seed,
)

data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)

trainer = Seq2SeqTrainer(
    model = model,
    args = training_args,
    train_dataset = train_dataset,
    optimizers = (get_optimizer(model, training_args, "AdamW"), None),
    data_collator = data_collator
)

trainer.train()

free_gpu()
```
