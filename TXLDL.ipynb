```python
import re
import string
```

```python
def normal_text(text):
    text = text.lower()
    text = re.sub('^\s', " ", text)
    dupSpace = r'\s\s+' 
    text = re.sub(dupSpace, " ", text) #fix duplicate space
    text = re.sub(r"\b(\w+)(?:\W\1\b)+",r"\1", text, flags=re.IGNORECASE) # fix duplicate word
    text = re.sub(r'(\w+)\1+', r'\1', text, flags=re. IGNORECASE)  # example : hahahahah -> haha 
    return text.strip()
```

```python
def clean_text(text):
    text = normal_text(text)
    return text
```

```python
for i in data_vi_raw:
    print(i, clean_text(i))
```
